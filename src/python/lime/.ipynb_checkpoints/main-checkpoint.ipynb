```python
import tensorflow as tf
slim = tf.contrib.slim
import sys
sys.path.append('/elena/models/tf-models/slim')
%load_ext autoreload
%autoreload 2
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
from nets import inception
from preprocessing import inception_preprocessing
```

```python
session = tf.Session()
```

```python
image_size = inception.inception_v3.default_image_size
def transform_img_fn(path_list):
    out = []
    for f in path_list:
        #import matplotlib.image as mpimg
        #img = mpimg.imread(f)
        image_contents = tf.read_file(f)
        image_raw = tf.image.decode_jpeg(image_contents, channels=3)
        #image_raw = tf.image.decode_jpeg(open(f).read(), channels=3)
        image = inception_preprocessing.preprocess_image(image_raw, image_size, image_size, is_training=False)
        out.append(image)
    return session.run([out])[0]
```

```python
from datasets import imagenet
names = imagenet.create_readable_names_for_imagenet_labels()
names
```

```python
processed_images = tf.placeholder(tf.float32, shape=(None, 299, 299, 3))
processed_images.shape
```

```python
import os
with slim.arg_scope(inception.inception_v3_arg_scope()):
    logits, _ = inception.inception_v3(processed_images, num_classes=1001, is_training=False)
probabilities = tf.nn.softmax(logits)

checkpoints_dir = '/elena/models/tf-models/slim/pretrained'
init_fn = slim.assign_from_checkpoint_fn(
    os.path.join(checkpoints_dir, 'inception_v3.ckpt'),
    slim.get_model_variables('InceptionV3'))
init_fn(session)
```

```python
def predict_fn(images):
    return session.run(probabilities, feed_dict={processed_images: images})
```

```python
images = transform_img_fn(["/elena/Google Drive/titech/research/experiments/input/py_tests/palm_interior.jpg"])
# I'm dividing by 2 and adding 0.5 because of how this Inception represents images
plt.imshow(images[0] / 2 + 0.5)
preds = predict_fn(images)
for x in preds.argsort()[0][-5:]:
    print(x, names[x], preds[0,x])
```

```python
image = images[0]
```

```python
from lime import lime_image
import time
```

```python
explainer = lime_image.LimeImageExplainer()
```

```python
tmp = time.time()
# Hide color is the color for a superpixel turned OFF. Alternatively, if it is NONE, the superpixel will be replaced by the average of its pixels
explanation = explainer.explain_instance(image.astype('double'), predict_fn, top_labels=5, hide_color=0, num_samples=1000)
print (time.time() - tmp)
```

## Insert sRGB of image with mole 

```python
basedir = "/elena/Google Drive/titech/research/experiments/input/py_tests/"

def prepareExplain(basedir, imgName):
    images = transform_img_fn([basedir + imgName + ".jpg"])
    # I'm dividing by 2 and adding 0.5 because of how this Inception represents images
    plt.imshow(images[0] / 2 + 0.5)
    preds = predict_fn(images)
    for x in preds.argsort()[0][-5:]:
        print(x, names[x], preds[0,x])
    image = images[0]
    explainer = lime_image.LimeImageExplainer()
    tmp = time.time()
    # Hide color is the color for a superpixel turned OFF. Alternatively, if it is NONE, the superpixel will be replaced by the average of its pixels
    explanation = explainer.explain_instance(image.astype('double'), predict_fn, top_labels=5, hide_color=0, num_samples=1000)
    print (time.time() - tmp)
    
    return explanation 
```

```python
imgName = "hand_mole"
explanation = prepareExplain(basedir, imgName)
```

```python
temp, mask = explanation.get_image_and_mask(45, positive_only=True, num_features=5, hide_rest=True)
plt.imshow(mark_boundaries(temp / 2 + 0.5, mask))
```

```python
temp, mask = explanation.get_image_and_mask(45, positive_only=False, num_features=10, hide_rest=False)
plt.imshow(mark_boundaries(temp / 2 + 0.5, mask))
```

```python
#Select the same class explained on the figures above.
ind =  explanation.top_labels[0]

#Map each explanation weight to the corresponding superpixel
dict_heatmap = dict(explanation.local_exp[ind])
heatmap = np.vectorize(dict_heatmap.get)(explanation.segments) 

#Plot. The visualization makes more sense if a symmetrical colorbar is used.
plt.imshow(heatmap, cmap = 'RdBu', vmin  = -heatmap.max(), vmax = heatmap.max())
plt.colorbar()
```
